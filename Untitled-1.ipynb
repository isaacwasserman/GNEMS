```python
from sklearn.metrics import v_measure_score
import numpy as np
import glob
import csv
import matplotlib.pyplot as plt
from skimage.color import label2rgb
from skimage.segmentation import slic, felzenszwalb, quickshift
```

```python
def score(gts, segmentation):
    scores = []
    for gt in gts:
        scores.append(v_measure_score(gt.flatten(), segmentation.flatten()))
    return np.mean(scores)

id = "388006"
image = plt.imread(f"datasets/BSDS500/BSDS500/data/images/test/{id}.jpg")
gt_paths = glob.glob(f"datasets/BSDS500/gt/{id}-*.csv")
gts = [np.genfromtxt(path, delimiter=',') for path in gt_paths]
plt.imshow(image)
plt.show()

slic_segmentation = slic(image, n_segments=25)
plt.imshow(label2rgb(slic_segmentation, image))
plt.title(f"SLIC: {score(gts, slic_segmentation)}")
plt.show()

quickshift_segmentation = quickshift(image)
plt.imshow(label2rgb(quickshift_segmentation, image))
plt.title(f"Quickshift: {score(gts, quickshift_segmentation)}")
plt.show()

felzenszwalb_segmentation = felzenszwalb(image)
plt.imshow(label2rgb(felzenszwalb_segmentation, image))
plt.title(f"Felzenszwalb: {score(gts, felzenszwalb_segmentation)}")
plt.show()
```
